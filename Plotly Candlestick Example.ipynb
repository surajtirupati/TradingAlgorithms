# Plotly Candlestick Example

```python
import plotly.graph_objects as go
import pandas as pd 
import yfinance as yf
```

```python
aapl = yf.Ticker("AAPL").history(period="30y")
aapl["MA"] = aapl.Close.rolling(20).mean()
```

```python
aapl = aapl[-252:]
```

```python
# initialise figure
fig = go.Figure()

# add candlestick chart
fig.add_trace(
    go.Candlestick(
        x=aapl.index,
        open=aapl['Open'],
        high=aapl['High'],
        low=aapl['Low'],
        close=aapl['Close'],
        name="Candlestick Chart"
                  
    )
)

# add scatter line
fig.add_trace(
    go.Scatter(
        x=aapl.index,
        y=aapl["MA"],
        line=dict(color="#4277f9"),
        name="20 Day Moving Average"
                  
    )
)

# configuring figure layout
fig.update_layout(
    title="AAPL Candlestick Chart",
    xaxis_title="Date",
    yaxis_title="Price",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
    
)

# display figure
fig.show()
```
