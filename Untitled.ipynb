```python
import yfinance as yf
import pandas as pd
import numpy as np
btc = yf.Ticker("BTC-USD").history(period="4y")
eth = yf.Ticker("ETH-USD").history(period="4y")
```

```python
def log_returns(df):
    """
    Description: Calculates log returns
    Takes input of dataframe containing price column

    Outputs: dataframe column containing log returns
    """
    log_price = np.log(df)
    log_returns = log_price- log_price.shift(1)
    log_returns = log_returns.dropna()
    return log_returns
```

```python
btc_log = log_returns(btc["Close"])
eth_log = log_returns(eth["Close"])
```

```python
returns = pd.concat([btc_log, eth_log], axis=1).reset_index()
returns.to_csv("returns.csv")
```

```python
np.corrcoef(btc_log, eth_log)
```

```python
(btc_log - eth_log).cumsum().plot()
```

```python
btc_vol = btc_log.std()
eth_vol = eth_log.std()
target_vol = 0.2
```

```python
vol_adj_btc = (target_vol/btc_vol) * btc_log
vol_adj_eth = (target_vol/eth_vol) * eth_log

vol_adj_btc.cumsum().plot()
```
